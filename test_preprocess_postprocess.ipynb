```python
import matplotlib.pyplot as plt
import torch

def plot_image(image, title=''):
    """
    Plot an image given as a PyTorch tensor or numpy array.

    Parameters:
        image (torch.Tensor or numpy.ndarray): Input image tensor or numpy array.
        title (str, optional): Title for the plot. Defaults to ''.
    """
    if isinstance(image, torch.Tensor):
        image = image.permute(1, 2, 0)  # Assuming tensor shape is (channels, height, width)
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()
```
